Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xp = np.array([0, 50])
yp = np.array([0, 100])

plt.plot(xp, yp)
plt.show()
```

<b>Plotting Without Line

```python
xp = np.array([0, 50])
yp = np.array([1, 10])

plt.plot(xp, yp, 'o')
plt.show()
```

<b>Multiple Points

```python
xp = np.array([1, 2, 4, 6, 10])
yp = np.array([4, 7, 2, 8, 6])

plt.plot(xp, yp)
plt.show()
```

<b>Default X-Points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, etc.

```python
yp = np.array([4, 2, 7, 5, 9, 7])

plt.plot(yp)
plt.show()
```

<b>Marker, Line, Color

```python
yp = np.array([4, 7, 2, 10])

plt.plot(yp, marker = 'o')
# plt.plot(yp, marker = '*', color = 'red')
# plt.plot(yp, marker = 'D', linestyle='dotted', color = '#008000')
# plt.plot(yp, marker = 'x', linestyle='dashdot', color = 'r')
# plt.plot(yp, marker = '+', ls='--', color = 'y')
# plt.plot(yp, marker = 'p', ls = '--')

plt.show()
```

<b>Format Strings using shortcut string notation parameter </b> - <font color=red> marker | line | color

```python
yp = np.array([3, 8, 1, 10])

plt.plot(yp, 'o:r')
# plt.plot(yp, '*:b')
# plt.plot(yp, '+--c')
# plt.plot(yp, 'd:k')

plt.show()
```

<b>Marker Size, Marker Color:</b>  ms, mec, mfc

```python
yp = np.array([3, 8, 1, 10])

plt.plot(yp, color = 'k', marker = 'o', ms = 15, mec = 'r', mfc = 'y')
plt.show()
```

<b>Line Width, Style, Color: </b> linewidth, linestyle (ls), color

```python
yp = np.array([4, 2, 7, 5, 9, 4])

plt.plot(yp, color = 'b', linewidth = '3', ls = '--', marker = 'o', ms = 10, mec = 'k', mfc = 'r')
plt.show()
```

<b>Multiple Lines

```python
yp1 = np.array([4, 8, 2, 9])
yp2 = np.array([5, 1, 9, 11])

plt.plot(yp1)
plt.plot(yp2)

plt.show()
```

```python
xp = np.array([1, 2, 4, 6, 10])
yp1 = np.array([4, 8, 2, 9, 7])
yp2 = np.array([5, 1, 7, 12, 9])

plt.plot(xp, yp1)
plt.plot(xp, yp2, color = 'g', linewidth = '1', ls = '-')

plt.show()
```

```python
x1 = np.array([0, 2, 6, 8])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([1, 3, 6, 7])
y2 = np.array([6, 2, 4, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

<b> Plot Title, Lables, Grid

```python
x = np.array([1990, 1995, 2000, 2005, 2010, 2015, 2020])
y = np.array([2410, 2700, 3232, 3288, 4280, 5290, 5500])

plt.plot(x, y)

plt.title("Village population growth")
plt.xlabel("Year")
plt.ylabel("Population")

plt.show()
```

```python
x = np.array([1990, 1995, 2000, 2005, 2010, 2015, 2020])
y = np.array([2410, 2700, 3232, 3288, 4280, 5290, 5500])

plt.plot(x, y)
plt.grid(axis = 'y', color = 'b', linestyle = '-.', linewidth = 0.2)

font1 = {'family':'verdana','color':'k','size':15}
font2 = {'family':'serif','color':'darkred','size':10}

plt.title("Village population growth", fontdict = font1, loc = 'center')
plt.xlabel("Year", fontdict = font2)
plt.ylabel("Population", fontdict = font2)

plt.show()
```

<b>Display Multiple Plots

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)
plt.plot(x,y)
plt.title("INCOME")

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2)
plt.plot(x,y)
plt.title("Sales")

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 3)
plt.plot(x,y)
plt.title("Employees")

plt.suptitle("MY SHOP")
plt.show()
```

<b>Scatter Plots

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,83,86,92,87,94,78,97,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
# plt.scatter(x, y, s=60, c='y', alpha=0.3)

plt.show()
```

<b> Bars

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2, 10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 8, 2, 10])

plt.barh(x, y, height = 0.7)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([4, 7, 2, 9])

plt.bar(x, y, color = 'r', width = 0.8)
plt.show()
```

<b>Histograms

```python
x = np.random.normal(170, 10, 250)

plt.hist(x, color = 'b')
plt.show() 
```

<b>Pie Charts

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["r", "hotpink", "b", "#4CAF50"]
myexplode = [0.1, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, colors = mycolors,  shadow = False)
plt.legend()
plt.show() 
```
